The of the scripts below is to prperly train 75% of our dataset and test the algorithm on the remaining 25% of the dataset with the aim of inferring the algorithm to nearly predict the type of clients for the rest of about 3000 customers whose staging have not yet been identified.

Our dataset was initially cleaned up from mySQL thus, there were no null variables in the csv file. The pandas package was useful in creating dummy variables for our predictive variable.

With a high training rate of 82% and a high testing rate of 80%, the algorithm was neither under-fitted nor over-fitted. The model and accuracy will not change even if our random seed was changed from 219 to 123 or 400. 

Based on the total dataset there were more categorical variables than quantitative which may be a slight limitation in leaving the model a bit biased.


```python
#Install the missing packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt # data visualization

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

#creating data frame for exporting SQL csv file
file = './SQL Export 2.csv'
retail = pd.read_csv(file)

#
retail.head(n=10)

#1
# taking the housing dataset and then transforming it into boolean based on if
#a value is null and then summing together the results per column
retail.isnull().any().any()
```

```python
#creating dummy variables for wholesaler and personal
dummy_variable = pd.get_dummies(retail['type_of_client'])

# joining codings together 
dummy_data = retail.join([dummy_variable])
dummy_data
```

```python
#3


# looping to make x-variables suitable for statsmodels
#creating y-variable and setting it to 0
retail['y'] = 0

#creating for loop to categorize wholesaler and personal
for index, value in retail.iterrows():
    if retail.loc[index,'type_of_client'] == 'Wholesaler':
        retail.loc[index,'y' ] = 1
    elif retail.loc[index, 'type_of_client'] == 'Personal':
        retail.loc[index, 'y'] = 0

#sets new identity to zero        
retail['identity_1'] = 0

#Creating for loop and setting males to 1
for index, value in retail.iterrows():
    if retail.loc[index,'identity'] == 'M':
        retail.loc[index,'identity_1' ] = 1
    else:
        retail.loc[index,'identity_1' ] = 0

#Setting employment type to 0 
retail['type_of_employment_1'] = 0

#creating for loop for type of employment and setting private to 1
for index, value in retail.iterrows():
    if retail.loc[index,'type_of_employment'] == 'Private':
        retail.loc[index,'type_of_employment_1' ] = 1
    else:
        retail.loc[index,'type_of_employment_1' ] = 0

#setting country to 1        
retail['country_1'] = 0

#creating country for United States and setting it to 1
for index, value in retail.iterrows():
    if retail.loc[index,'country'] == 'United-States':
        retail.loc[index,'country_1' ] = 1
    else:
        retail.loc[index,'country_1' ] = 0

#setting education to 0        
retail['education_1'] = 0        
  
#Creating for loop where education is set to 1 for HS Graduate    
for index, value in retail.iterrows():
    if retail.loc[index,'education'] == 'HS-grad':
        retail.loc[index,'education_1' ] = 1
    else:
        retail.loc[index,'education_1' ] = 0   
        
#Setting marital status to 0        
retail['marital_status_1'] = 0        
 
#Creating for loop with marital status and selecting Married to 1    
for index, value in retail.iterrows():
    if retail.loc[index,'marital_status'] == 'Married-civ-spouse':
        retail.loc[index,'marital_status_1' ] = 1
    else:
        retail.loc[index,'marital_status_1' ] = 0       

#setting household relationship to 0
retail['household_relationship_1'] = 0        
 
#creating for loop with Husband set to 1    
for index, value in retail.iterrows():
    if retail.loc[index,'household_relationship'] == 'Husband':
        retail.loc[index,'household_relationship_1' ] = 1
    else:
        retail.loc[index,'household_relationship_1' ] = 0       


retail.head()

retail
```

```python
#4
# preparing explanatory variable data
retail_data = retail.drop(['customer',
                               'type_of_client',
                               'identity',
                               'birthdate',
                               'current_age',
                               'occupation_type',
                               'type_of_employment',
                               'education',
                               'marital_status',
                               'household_relationship',
                               'country',
                               'y'],
                                axis = 1)


# preparing response variable data
retail_target = retail.loc[ : , 'y']
log_retail_target = retail.loc[ : , 'y'] # ready for use later


# preparing training and testing sets (all letters are lowercase)
x_train, x_test, y_train, y_test = train_test_split(
            retail_data,
            retail_target,
            test_size = 0.25,
            random_state = 219)


# checking the shapes of the datasets
print(f"""
Training Data
-------------
X-side: {x_train.shape}
y-side: {y_train.shape[0]}



Testing Data
------------
X-side: {x_test.shape}
y-side: {y_test.shape[0]}
""")
```

```python
#5
# preparing explanatory variable data
retail_data = retail.drop(['customer',
                               'type_of_client',
                               'identity',
                               'birthdate',
                               'current_age',
                               'occupation_type',
                               'type_of_employment',
                               'education',
                               'marital_status',
                               'household_relationship',
                               'country',
                               'y'],
                                axis = 1)


# preparing response variable data
retail_target = retail.loc[ : , 'y']

#Setting train and testing model
x_train, x_test, y_train, y_test = train_test_split(retail_data, retail_target
                            , test_size=0.25, random_state=219)
#creating variable for logistic regression
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
 
#Creating variable for predictions to see how well it fits    
predictions = logmodel.predict(x_test)

#Prints out classification report, confusion matrix, and accuracy
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))

#Saves the train score for further analysis 
train_score = print(accuracy_score(y_test, predictions))
```

```python
# preparing response variable data 
retail_target = retail.loc[ : , 'y']

#Same as the one above but test size equals to .75
x_train, x_test, y_train, y_test = train_test_split(retail_data, retail_target
                    , test_size=0.75, random_state=219)

#creates models for logistic Regression
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
predictions = logmodel.predict(x_test)

#Prints out classification report, confusion matrix, and accuracy score
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))

#Saves the test score for further analysis 
test_score = print(accuracy_score(y_test, predictions))

```
